```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train), (X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255
X_test = X_test/255  #bring all values in the range of 0 to 1
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28, 28))) # brings the 2d data to 1d OCnverts 28x28 2D data to 784
model.add(Dense(128, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(10, activation='softmax')) #Output layer (Multiclass problem therefore we use 10 nodes in the output layer)
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
```

```python
model.fit(X_train, y_train, epochs=50, validation_split=0.2)
```

```python
 y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
